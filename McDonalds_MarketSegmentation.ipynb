```python
#categorical data
from sklearn.preprocessing import LabelEncoder
def labelling(x):
    data[x] = LabelEncoder().fit_transform(data[x])
    return data

cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for i in cat:
    labelling(i)
data
```

```python
data_eleven = data.loc[:,cat]
data_eleven
```

```python
plt.rcParams['figure.figsize'] = (11,14)
data.hist()
plt.show()
```

```python
x = data.loc[:,cat].values
x
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca_data = preprocessing.scale(x)

pca = PCA(n_components=11)
pc = pca.fit_transform(x)
names = ['p1','p2','p3','p4','p5','p6','p7','p8','p9','p10','p11']
pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["P"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = data_eleven.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
#K-means clustering 

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(data_eleven)
data['cluster_num'] = kmeans.labels_ 
print (kmeans.labels_) 
print (kmeans.inertia_) 
print(kmeans.n_iter_) 
print(kmeans.cluster_centers_) 
```

```python
data['VisitFrequency'] = LabelEncoder().fit_transform(data['VisitFrequency'])
visit = data.groupby('cluster_num')['VisitFrequency'].mean()
visit = visit.to_frame().reset_index()
visit
```

```python
data['Gender'] = LabelEncoder().fit_transform(data['Gender'])
Gender = data.groupby('cluster_num')['Gender'].mean()
Gender = Gender.to_frame().reset_index()
Gender
```

```python
data['Like'] = LabelEncoder().fit_transform(data['Like'])
Like = data.groupby('cluster_num')['Like'].mean()
Like = Like.to_frame().reset_index()
Like
```

```python
segment = Gender.merge(Like, on='cluster_num', how='left').merge(visit, on='cluster_num', how='left')
segment
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = "VisitFrequency", y = "Like",data=segment,s=400, color="skyblue")
plt.title("Segmentation evaluation plot for the fast food data set",
          fontsize = 15) 
plt.xlabel("Visit to mcd", fontsize = 15) 
plt.ylabel("Likes", fontsize = 15) 
plt.show()
     
```
